```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization,Dropout, Input
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.regularizers import l1_l2
from tensorflow.keras.layers import LeakyReLU, PReLU
from tensorflow.keras.callbacks import ReduceLROnPlateau
from tensorflow.keras.constraints import MaxNorm
```

```python
# Load and prepare the data
df = pd.read_csv('cleaned_stock_data.csv')
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')
df.head()
```

```python
# Calculate technical indicators
def add_technical_indicators(df, stock='AMZN'):
    # Select price and volume columns
    price_col = f'Adj Close_{stock}'
    volume_col = f'Volume_{stock}'
    
    # Moving averages
    df[f'MA5_{stock}'] = df[price_col].rolling(window=5).mean()
    df[f'MA20_{stock}'] = df[price_col].rolling(window=20).mean()
    
    # Price momentum
    df[f'Return_5d_{stock}'] = df[price_col].pct_change(5)
    df[f'Return_20d_{stock}'] = df[price_col].pct_change(20)
    
    # Volatility
    df[f'Volatility_{stock}'] = df[price_col].pct_change().rolling(window=10).std()
    
    # RSI (Relative Strength Index)
    delta = df[price_col].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
    rs = gain / loss
    df[f'RSI_{stock}'] = 100 - (100 / (1 + rs))
    
    # Volume indicators
    df[f'Volume_MA20_{stock}'] = df[volume_col].rolling(window=20).mean()
    df[f'Volume_Ratio_{stock}'] = df[volume_col] / df[f'Volume_MA20_{stock}']
    
    return df
```

```python
# Add technical indicators
df = add_technical_indicators(df, 'AMZN')
df.head()
```

```python
# Create target variable (1 for buy, 0 for sell)
# Buy signal if price increases by 1% or more in next 5 days
future_returns = df['Adj Close_AMZN'].shift(-5) / df['Adj Close_AMZN'] - 1
df['Target'] = (future_returns > 0.01).astype(int)
```

```python
# Select features for the model
feature_columns = [
    'Adj Close_AMZN', 'Volume_AMZN',
    'MA5_AMZN', 'MA20_AMZN',
    'Return_5d_AMZN', 'Return_20d_AMZN',
    'Volatility_AMZN', 'RSI_AMZN',
    'Volume_MA20_AMZN', 'Volume_Ratio_AMZN'
]
```

```python
# Drop rows with NaN values
df = df.dropna()
df.head()
```

```python
# Prepare features and target
X = df[feature_columns].values
y = df['Target'].values

# Scale the features
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, shuffle=False)
```

```python
# Create the Dense Neural Network model
#input_dim = len(feature_columns)

model = Sequential([
        Input(shape=(len(feature_columns),)),
        # Input layer with regularization
        Dense(64, 
              activation='relu',
              kernel_regularizer=l1_l2(l1=1e-5, l2=1e-4),
              bias_regularizer=l1_l2(l1=1e-4, l2=1e-3),
              activity_regularizer=l1_l2(l1=1e-4, l2=1e-3),
              kernel_constraint=MaxNorm(max_value=3)),
              BatchNormalization(),
        Dropout(0.4),
        
        # Bottleneck layer
        Dense(32,
              activation='relu',
              kernel_regularizer=l1_l2(l1=1e-5, l2=1e-4),
              bias_regularizer=l1_l2(l1=1e-4, l2=1e-3),
              kernel_constraint=MaxNorm(max_value=3)),
        BatchNormalization(),
        Dropout(0.3),
        
        # Expansion layer
        Dense(48,
              activation='relu',
              kernel_regularizer=l1_l2(l1=1e-5, l2=1e-4),
              kernel_constraint=MaxNorm(max_value=3)),
        BatchNormalization(),
        Dropout(0.2),
        
        # Output layer
        Dense(1, activation='sigmoid', kernel_regularizer=l1_l2(l1=1e-5, l2=1e-4))
    ])
```

```python
# Compile the model

# Custom learning rate schedule
initial_learning_rate = 0.001
decay_steps = 1000
decay_rate = 0.9

# Compile model with fixed learning rate
optimizer = tf.keras.optimizers.Adam(
    learning_rate=0.001,  # Fixed initial learning rate
    beta_1=0.9,
    beta_2=0.999,
    epsilon=1e-07,
    amsgrad=True
)

model.compile(
    optimizer=optimizer,
    loss='binary_crossentropy',
    metrics=[
        'accuracy',
        tf.keras.metrics.AUC(),
        tf.keras.metrics.Precision(),
        tf.keras.metrics.Recall()
    ]
)

# Print model summary
model.summary()
```

```python
# Train the model
class_weights = {
    0: len(y_train) / (2 * np.sum(y_train == 0)),
    1: len(y_train) / (2 * np.sum(y_train == 1))
}

history = model.fit(
    X_train, y_train,
    epochs=150,
    batch_size=32,
    validation_split=0.2,class_weight=class_weights,
    callbacks = [
        # Early stopping
        tf.keras.callbacks.EarlyStopping(
            monitor='val_loss',
            patience=15,
            restore_best_weights=True
        ),
        # Learning rate reduction
        tf.keras.callbacks.ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=5,
            min_lr=1e-6
        )
    ],
    verbose=1
)
```

```python
# Plot training history
plt.figure(figsize=(12, 4))

# Plot accuracy
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training')
plt.plot(history.history['val_accuracy'], label='Validation')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Plot loss
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training')
plt.plot(history.history['val_loss'], label='Validation')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Make predictions
train_pred = model.predict(X_train)
test_pred = model.predict(X_test)

# Convert predictions to binary (0 or 1)
train_pred_binary = (train_pred > 0.5).astype(int)
test_pred_binary = (test_pred > 0.5).astype(int)
```

```python
# Print classification reports
print("\nTraining Set Classification Report:")
print(classification_report(y_train, train_pred_binary,zero_division=0))

print("\nTest Set Classification Report:")
print(classification_report(y_test, test_pred_binary,zero_division=0))
```

```python
# Plot confusion matrix
plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_test, test_pred_binary)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix (Test Set)')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
```
